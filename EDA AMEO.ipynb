* Since it is clearly visible that there are outliers so we are removing them by using IQR method in the next step

```python
def outliers(df,x):
    q1= df[x].quantile(0.25)
    q3= df[x].quantile(0.75)
    iqr= q3-q1
    lower_bound= q1-1.5*iqr
    upper_bound= q3+1.5*iqr
    #df[x]= df[x]
    df2 = df[(df[x] >= lower_bound) & (df[x] <= upper_bound)]
    df=df2
    return df
```

**This above fuction will work for us to remove all outiers from the respected features**

```python
df2 = outliers(df,'Salary')
```

# Reasearch Q2

### Is there a relationship between gender and specialisation? (i.e. Does the preference of Specialisation depend on the Gender?)


To find this, we can perform some tests such as "chi" test which can give us exact conclusion about this question

```python
from scipy.stats import chi2_contingency
```

```python
# Create a contingency table
contingency_table = pd.crosstab(df5['Gender'], df5['Specialization'])

# Perform chi-square test for independence
chi2, pval, dof, expected = chi2_contingency(contingency_table)

# Print the results
print(f"Chi-square statistic: {chi2}")
print(f"P-value: {pval}")
if pval < 0.05:
    print("There is a significant relationship between gender and specialization.")
else:
    print("There is no significant relationship between gender and specialization.")
```

Since this test has got value more than 0.05, it is clear that there are relationship between gender and specialization.
